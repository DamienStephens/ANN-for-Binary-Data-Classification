This file shows a simple example of binary data classification using an artificial neural network (ANN).

# Imports

```python
import tensorflow as tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import confusion_matrix, accuracy_score
```

# Data Preprocessing and Exploration

```python
# import the dataset from Churn_Modelling.csv
data_path = 'Churn_Modelling.csv'
dataset = pd.read_csv(data_path)
```

```python
dataset.head()
```

```python
# select the independent variable
# use drop() to select the matrix of features (the independent variables) by dropping columns that are not
# correlated with the output
# axis=0 selects the indexes, axis=1 selects the columns
x = dataset.drop(labels=['RowNumber', 'CustomerId', 'Surname', 'Exited'], axis=1)

# select the dependent variable
y = dataset['Exited']
```

```python
# Geography has 3 categories and Gender has 2 categories
# we will convert this categorical data into numerical data below
x.head(10)
```

```python
y.head()
```

```python
# encode the Geography categorical data
# fit the data into the object label_1, then transform the data
label_1 = LabelEncoder()
x['Geography'] = label_1.fit_transform(x['Geography'])
```

```python
# now the Geography column has been converted into numerical data
x.head(10)
```

```python
# encode the Gender categorical data
label_2 = LabelEncoder()
x['Gender'] = label_2.fit_transform(x['Gender'])
```

```python
# now the Gender column has been converted into numerical data
x.head(10)
```

```python
# avoid the "dummy variable trap"
# this splits the Geography column into two columns where the original geography has been changed to 
# binary (00, 01, 10, 11) so the model does not interpret one geographic location higher than another
x = pd.get_dummies(data=x, drop_first=True, columns=['Geography'])
```

```python
x.head(10)
```

```python
# split the dataset into train and test datasets
# test size = 20% of the dataset, so training size is 80% of the dataset
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)
```

```python
# feature scaling
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
x_train
```

```python
x_test
```

# Build the Model

```python
# create an object (initializing the ANN)
model = tf.keras.models.Sequential()
```

```python
# check the input_dimensions
x.shape
```

```python
# add the input layer
# units=6     trick: take average of input=11 and output=1, hence (11+1)/2=6
# input dimension=11
model.add(tf.keras.layers.Dense(units=6, activation='relu', input_dim=11))
```

```python
# add the hidden layer
model.add(tf.keras.layers.Dense(units=6, activation='relu'))
```

```python
# add the output layer
# since we have binary outputs, use sigmoid instead of softmax
model.add(tf.keras.layers.Dense(units=1, activation='sigmoid'))
```

```python
model.summary()
```

```python
# compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

# Train the Model

```python
# train the ANN model
model.fit(x_train, y_train.to_numpy(), batch_size=10, epochs=20)
```

# Model Evaluation and Prediction

```python
# evaluate the model
test_loss, test_accuracy = model.evaluate(x_test, y_test.to_numpy())
```

```python
print(f"Test Accuracy: {test_accuracy*100:.2f}%")
```

```python
# make predictions
y_pred = model.predict_classes(x_test)
```

```python
# look at the predicted values
print(y_pred)
```

```python
# look at the actual values
print(y_test)
```

```python
# convert the vector to an array
y_test = y_test.to_numpy()
```

```python
# make individual predictions based on index location
index = 0
print(f"y_pred[{index}]: {y_pred[index]}") 
print(f"y_test[{index}]: {y_test[index]}")
```

```python
# confusion matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```

```python
# calculate accuracy based on confusion matrix
acc_cm = accuracy_score(y_test, y_pred)
print(f"Accuracy of Confusion Matrix: {acc_cm*100:.2f}%")
```
